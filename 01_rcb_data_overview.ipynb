```python
# Import Libraries
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the datasets
```

```python
matches = pd.read_csv("matches.csv")
deliveries = pd.read_csv("deliveries.csv")

print("Shape of the matches dataset: ",matches.shape)
print("Shape of the deliveries dataset: ",deliveries.shape)

```

```python
# Quick sanity check
```

```python
print("Shape of the matches dataset: ",matches.shape)

print("Shape of the deliveries dataset: ",deliveries.shape)

print("First 5 rows of the matches dataset: ",matches.head())

print("\nAll columns in matches dataset:", matches.columns.tolist())
```

```python
# Check whether any single franchise with different names
```

```python
pd.concat([matches['team1'], matches['team2']]).unique() 
```

```python
# As we can see a franchise with two names: Royal challengers Bangalore and Royal challenegers bengaluru, we replace it with single name
```

```python
matches.replace(
    {
        'Royal Challengers Bangalore': 'Royal Challengers Bengaluru'
    },
    inplace=True
)
```

```python
# Select the team and Filter the team matches
```

```python
My_Team = 'Royal Challengers Bengaluru'

rcb_matches = matches[ (matches['team1'] == My_Team) | (matches['team2'] == My_Team )]

print(f"Total IPL matches in dataset: {len(matches)}")
print(f"Total no of matches played by RCB: {len(rcb_matches)}")
print(f"RCB has played in {(len(rcb_matches)/len(matches))*100 :.1f}% of all IPL matches")
```

```python
#Quick look at their most recent matches
```

```python
print(rcb_matches[['season', 'team1', 'team2', 'winner', 'venue']].tail(5))
```

```python
# Data Quality Check
```

```python
# Check missing values in RCB's match data
missing_values = rcb_matches.isnull().sum()
missing_percentage = missing_values / len(rcb_matches)

# Create a summary dataframe
missing_df = pd.DataFrame({
            'missing_values' : missing_values,
            'missing_percentage' : missing_percentage
})

# Show only columns that have missing value
missing_df = missing_df[missing_df['missing_values'] > 1]
if len(missing_df) > 1:
    print(f"Columns with missing values in RCB match Data: {missing_df.sort_values(by = 'missing_values', ascending = False)}")
else:
    print("No missing values found in the data")
```

```python
# RCB's match analysis
```

```python
# Calculate win percentage
rcb_wins = len(rcb_matches[rcb_matches['winner'] == My_Team])
rcb_losses = len(rcb_matches) - rcb_wins
win_percentage = (rcb_wins / len(rcb_matches)) * 100

print(f"RCB Wins: {rcb_wins}")
print(f"RCB Losses: {rcb_losses}")
print(f"Win Percentage: {win_percentage:.1f}%")
```
