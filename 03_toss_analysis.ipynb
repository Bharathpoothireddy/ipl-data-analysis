```python
# Import Libraries
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load Data
```

```python
rcb_matches = pd.read_csv(r"rcb_clean.csv")
matches = pd.read_csv(r"matches.csv")
My_team = "Royal Challengers Bengaluru"

print(f"RCB matches loaded:{rcb_matches.shape[0]} rows")
rcb_matches.head(2)
```

```python
# WIN% WHEN RCB WINS THE TOSS
```

```python
rcb_won_toss = rcb_matches[rcb_matches['toss_winner'] == My_team]
rcb_won_match = rcb_won_toss[rcb_won_toss['winner'] == My_team]
toss_won_match_won = round((len(rcb_won_match) / len(rcb_won_toss)) * 100, 1)

print(f" When RCB wins toss: Their win% = {toss_won_match_won}%")
```

```python
# WIN% WHEN RCB LOSS THE TOSS
```

```python
rcb_lost_toss = rcb_matches[rcb_matches['toss_winner'] != My_team]
rcb_won_match = rcb_lost_toss[rcb_lost_toss['winner'] == My_team]
toss_lost_match_won = round((len(rcb_won_match) / len(rcb_lost_toss)) * 100, 1)
print(f" When RCB lose toss: Their win% = {toss_lost_match_won}%")
```

```python
# BAR CHART: TOSS vs MATCH OUTCOME
```

```python
labels = ['Won Toss', 'Lost Toss']
values = [toss_won_match_won, toss_lost_match_won]
plt.bar(labels, values, color=['green', 'red'], edgecolor='black', width=0.4)
plt.grid(axis='y')
plt.axhline(y=48.2, linestyle="-",color='blue', label="Overall Average win rate(48.2)")
plt.ylabel("Match Win %", fontsize=12)
plt.title("RCB: Toss vs Match Outcome", fontsize=14, fontweight='bold')
plt.legend()
plt.tight_layout()
plt.show()
```

Day 3 Summary: Toss Impact Analysis

Key Finding
- **Toss won → Win%:** 50.4%
- **Toss lost → Win%:** 46.3%
- **Overall average:** 48.2%

Conclusion
Winning the toss slightly improves RCB's chances of winning. The difference is only 4.1 percentage points — much smaller than the common "toss decides the match" narrative suggests.
